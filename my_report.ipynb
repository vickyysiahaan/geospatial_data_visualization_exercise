### 2. Data Preparation

The shipping address values are not in a standard format. Information of street name, kelurahan, kecamatan, city, etc. are stored in unstructured way. They are in one column. So we need to transform the value. People usually write an address in this common format:

*street name and number, kelurahan, kecamatan, city, province*

They are separated by commas. So let's try to extract information from some sample of address by splitting the data address by comma separator.

```python
sample_address = df["to_address"].sample(5).values
result = []
for i, item in enumerate(sample_address):
    result.append(item.split(","))
    print(result, len(result))
```

From the sample above we can see that many of them do not follow the common format. Some of them are even misspelled. Hence, it is difficult to use the full address information for shipment destination segmentation. It is not an easy task to correct large number of address data one by one. For this analysis, let's just use the last item of the extracted address info, which is hopefully the city name.

```python
# function to get city name
# assumption: all city name are maximum in two words.
# example: + Jakarta Selatan
#          + Bekasi
#          + Banda Aceh
# city with word length more than 2 will be filled with nan value (invalid)
def get_city_name(address_list):
    result = []
    for each in address_list:
        city_name = each.split(",")[-1]
        words = city_name.split(" ")
        try: 
            words.remove("")
        except:
            pass
        word_count = len(words)
        if word_count > 2:
            city_name = np.nan
        result.append(city_name)
    return result


df["city"] = get_city_name(df["to_address"].values)
df.head()
```

Remove data with invalid city name (NaN)

```python
df = df.dropna() 
df.head()
```

It is easier for us to segmentate geographical area by using longitude and latitude data. Lets get the longitude and latitude data of the city. We are using *Free* Nominatim API provided in geopy module for geocoding here.

```python
#uncomment this line below
#conda install -c conda-forge geopy
```

```python
from geopy.geocoders import Nominatim
from geopy.exc import GeocoderTimedOut
import time

geolocator = Nominatim(user_agent="test")

#function to get longitude and latitude of an address
def get_long_lat(city_list):
    result = []
    max_trial = 1
    for i, each in enumerate(city_list):        
        trial = 0
        while(trial <max_trial): 
            try:
                location = geolocator.geocode(each, timeout=5)
                result.append([location.longitude, location.latitude])
                break
            except:
                trial += 1
        if trial >= max_trial:
            result.append([np.nan, np.nan])
        time.sleep(1)
    longitude, latitude = zip(*result)
    return longitude, latitude
```

Based on <a href="https://operations.osmfoundation.org/policies/nominatim/"> the Nominatim usage policy</a>, heavy uses are not provided. This API can only handle an absolute maximum of 1 request per second per client. We are now having 812671 records of data.  It would take approximately 812671 / 3600 = 225.74 hours to convert all address data to longitude and latitude data. Therefore, we need to sample the dataset. Let's just take 500 samples of the data for this shipment destination segmentation to save time.  


```python
sample_set = df.sample(5000)
sample_set.head()
```

```python
t0 = time.time()
result = get_long_lat(sample_set["city"])
t1 = time.time()
print("processing time: ", t1-t0/3600, " hours")

# Add longitude and latitude value to dataframe
sample_set["longitude"]= result[0]
sample_set["latitude"]= result[1]
sample_set.head()

# Drop NaN DATA
sample_set = sample_set.dropna()
print(sample_set.shape)

# Save sample data
sample_set.to_csv("5000_samples.csv")

# sample_set = pd.read_csv("5000_samples.csv")
# sample_set.head()
```

### 3. Modelling

Because we have no prior knowledge about how many cluster of shipment destination we have, I think Density Based Clustering Algorithm (DBSCAN) is a good solution. This algorithm doesn't  

```python
from sklearn.cluster import DBSCAN
import sklearn.utils
from sklearn.preprocessing import StandardScaler
sklearn.utils.check_random_state(1000)

Clus_dataSet = pdf[['xm','ym']]
Clus_dataSet = np.nan_to_num(Clus_dataSet)
Clus_dataSet = StandardScaler().fit_transform(Clus_dataSet)

# Compute DBSCAN
db = DBSCAN(eps=0.15, min_samples=10).fit(Clus_dataSet)
core_samples_mask = np.zeros_like(db.labels_, dtype=bool)
core_samples_mask[db.core_sample_indices_] = True
labels = db.labels_
pdf["Clus_Db"]=labels

realClusterNum=len(set(labels)) - (1 if -1 in labels else 0)
clusterNum = len(set(labels)) 


# A sample of clusters
print(pdf.shape)
pdf[["to_address", "city" ,"longitude", "latitude", "Clus_Db"]].head(5)
```

```python
from mpl_toolkits.basemap import Basemap
import matplotlib.pyplot as plt
from pylab import rcParams
%matplotlib inline
rcParams['figure.figsize'] = (18,12)

my_map = Basemap(projection='merc',
            resolution = 'l', area_thresh = 1000.0,
            llcrnrlon=llon, llcrnrlat=llat, #min longitude (llcrnrlon) and latitude (llcrnrlat)
            urcrnrlon=ulon, urcrnrlat=ulat) #max longitude (urcrnrlon) and latitude (urcrnrlat)

my_map.drawcoastlines()
my_map.drawcountries()
#my_map.drawmapboundary()
my_map.fillcontinents(color = 'white', alpha = 0.3)
my_map.shadedrelief()

# To create a color map
colors = plt.get_cmap('jet')(np.linspace(0.0, 1.0, clusterNum))

#Visualization1
for clust_number in set(labels):
    c=(([0.4,0.4,0.4]) if clust_number == -1 else colors[np.int(clust_number)])
    clust_set = pdf[pdf.Clus_Db == clust_number]                    
    my_map.scatter(clust_set.xm, clust_set.ym, color =c,  marker='o', s= 20, alpha = 0.85)
    if clust_number != -1:
        cenx=np.mean(clust_set.xm) 
        ceny=np.mean(clust_set.ym) 
        
        cenLong = np.mean(clust_set.longitude)
        cenLat = np.mean(clust_set.latitude)
        cenLocation = geolocator.reverse(cenLat, cenLong).address
        clusterSize = len(pdf[pdf["Clus_Db"] == clust_number].values)
        plt.text(cenx,ceny,str(clust_number), fontsize=25, color='red',)
        print ("Cluster "+str(clust_number)+', Centroid longitude: '+ str(cenLong)+',  Centroid latitude: '+ str(cenLat)+',  Centroid_location: '+ str(cenLocation)+',  Cluster_Size: '+ str(clusterSize))
```

### 4. Analysis
Based on the result we get from Density Based Clustering, we have 5 segments of shipment destination in the period. The centre of each cluster is the centroid longitude and centroid latitude. Unfortunately, we can't get the centroid location (address) because of the limitation in the Nominatim API. But, we can visually see the cluster location by seeing the map above. Generally, the largest number of shipment is in the cluster 0, which has 144 points out of 5000 samples we have. We can see that cluster 0 are located arround Banten and West Java. So, this area has the highest shipment traffic. We can use the same analysis method for the other clusters.
